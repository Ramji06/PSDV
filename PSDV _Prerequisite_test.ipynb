```python
# Create a list of the first 20 natural numbers.
# Compute the sum of all even numbers in the list.
# Compute the product of all odd numbers in the list.
# Calculate the mean, median, and standard deviation of the list.

n= 20
nums= []               
sum_even=0
sum_odd=0

for i in range(1, n+1):
    nums.append(i)          
    if(i%2==0): 
        sum_even +=i
    else:
        sum_odd +=i

mean= (sum_even+sum_odd)/n
median= -1                                    #intializing median
if(n%2==0):                                   # as the list if already sorted therefore if n is even then averaging the middle 2 nums in the list 
    median= (nums[n//2-1]+ nums[n//2])/2 
else:                                         # taking the middle value if n is odd 
    median= nums[n//2]
stdDeviation= 0                               # using formula: std deviation=  root of (sum of (el- mean)*)
for el in nums:
    stdDeviation+= (el-mean)**2
stdDeviation/=n
stdDeviation**= 0.5

print(nums)
print(sum_even)
print(sum_odd)
print(mean)
print(median)
print(stdDeviation)
```

```python
# Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
# Plot the Fibonacci sequence using a line graph.

import matplotlib.pyplot as plt

n= int(input())
fibonacciSeies= [0, 1]
index= [1, 2]
for i in range(2, n):
    x= fibonacciSeies[-1]+fibonacciSeies[-2]
    fibonacciSeies.append(x)
    index.append(i+1)
print(fibonacciSeies)

plt.plot(index, fibonacciSeies)
plt.show
```

```python
# Write a program to check if a string is a palindrome (ignoring spaces and case).
# Input: "A man a plan a canal Panama"
# Output: True

s= input()
s= s.lower()
st= ""
for el in s:
    if(ord(el) in range(ord('a'), ord('a')+27)): 
        st+= el
if(st==st[::-1]):
    print(True)
else:
    print("False")
```

```python
# Create a dictionary of student names (keys) and their marks (values).
# Compute the average marks.
# Print the names of students scoring above the average.
# Visualize the distribution of marks using a bar chart.

import matplotlib.pyplot as plt

marks={
    "Ramji": 85,
    "John": 72,
    "Aaradhya": 80,
}
x= list(marks.keys())
y=[]
n= len(marks)
sum=0
for name in marks:
    sum+= marks[name]
    y.append(marks[name])
avgMarks= sum/ n
print(avgMarks)
for name in marks:
    if(marks[name]>avgMarks):
        print(name)

plt.bar(x,y, width=.2)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
# Given a list of strings, write a Python function to:
# Find the longest string.
# Count how many strings start with a vowel.
# Visualize the distribution of string lengths using a histogram.

import matplotlib.pyplot as plt

l= list(map(str, input().strip().split()))
strLen=[]
longest= l[0]
vowels=['a','e','i','o','u', 'A', 'B', 'C', 'D', 'E']
cnt=0                                     #count of string starting with vowels
for st in l:
    strLen.append(len(st))
    if(len(st)> len(longest)):
        longest= st
    if(st[0] in vowels): 
        cnt+=1

print(longest)
print(cnt)

w=[]
for i in range(1, len(longest)+1):
    w.append(i)

plt.hist(strLen, bins=w, edgecolor='r')
plt.show()
```

```python
# Write a program to:
# Split a string into a list of words.
# Remove all words that are shorter than 4 characters.
# Sort the remaining words alphabetically.
# Visualize the word lengths using a bar chart.

import matplotlib.pyplot as plt

s= input()
s=s.lower()
l= s.split()

words=[]           # list to store words greater than letter equal 4
for el in l:
    if(len(el)>=4):
        words.append(el)

words.sort()
wordLen=[]

for st in words:
    wordLen.append(len(st))
    
plt.bar(words, wordLen, width=0.4)
plt.xlabel("Words")
plt.ylabel("Len of Words")
plt.xticks(rotation=90)
plt.show()
```

```python
# Write a Python program that takes a numerical grade (0-100) as input and prints the corresponding letter grade (A, B, C, D, F).

def grade(n):
    if(n>=90):
        return 'A'
    elif(n>=80):
        return 'B'
    elif(n>=70):
        return 'C'
    elif(n>=50):
        return 'D'
    else:
        return 'F'

n= int(input())
print(grade(n))
```
